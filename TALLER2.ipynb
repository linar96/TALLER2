3. Numeros primos 2:
Escriba una funci√≥n que devuelva una lista de numeros primos con base a la cantidad que el usuario le solicite, Ejemplo si el usuario ingresa 3 debe retornar los 3 primeros numeros primos.


```python
nump=int(input("Ingrese la cantidad de primos que desea obtener : "))
def primo():
  numero=2
  yield numero
  while True:
    num = numero
    while True:
      num +=1
      con=1
      condiv=0

      while con<=num:
         if num % con ==0:
            condiv +=1
         if condiv >2:
            break
         con +=1
      if condiv ==2:
         yield num 
  
num_primo=primo()
lpri=[next(num_primo) for _ in range(nump)]
print(lpri)
```

6.  Numeros romanos:
Escriba una funci√≥n que reciba un numero y retorne como resultado el numero romano de dicho n√∫mero.

```python
n=int(input("Ingrese el numero a convertir: "))
num_map = [(1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'), (100, 'C'), (90, 'XC'),
           (50, 'L'), (40, 'XL'), (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')]


def conv(num):

    roman = ''

    while num > 0:
        for i, r in num_map:
            while num >= i:
                roman += r
                num -= i

    return roman
print(conv(n))
```
